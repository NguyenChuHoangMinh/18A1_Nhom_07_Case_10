2.0 - Chuẩn bị dữ liệu

```python
import pandas as pd
duong_dan = '../data_clean/'

df_thong_tin = pd.read_csv(duong_dan + 'sensor_info_clean.csv')
df_doc_cam_bien = pd.read_csv(duong_dan + 'sensor_readings_clean.csv')
df_bao_tri = pd.read_csv(duong_dan + 'maintenance_log_clean.csv')

df_thong_tin.head()
```

2.1 - Thống kê số lượng cảm biến theo từng loại (sensor_type)

```python
so_luong_theo_loai = df_thong_tin['sensor_type'].value_counts()
so_luong_theo_loai
```

2.2 - Tìm các bản ghi readings có value hoặc unit bất thường

```python
mask_value_nan = df_doc_cam_bien['value'].isna()
mask_unit_sai = ~df_doc_cam_bien['unit'].isin(['°C', 'mm/s', 'A'])

df_bat_thuong = df_doc_cam_bien[mask_value_nan | mask_unit_sai]
df_bat_thuong.head()
```

2.3 - Thống kê số lần bảo trì theo từng machine_id

```python
so_lan_bao_tri = df_bao_tri['machine_id'].value_counts()
so_lan_bao_tri
```
