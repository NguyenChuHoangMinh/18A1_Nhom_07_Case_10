4.0 - Chuẩn bị dữ liệu

```python
import pandas as pd
duong_dan = '../data_clean/'

df_thong_tin = pd.read_csv(duong_dan + 'sensor_info_clean.csv')
df_doc_cam_bien = pd.read_csv(duong_dan + 'sensor_readings_clean.csv')
df_bao_tri = pd.read_csv(duong_dan + 'maintenance_log_clean.csv')
```

4.1 - Merge sensor_info với sensor_readings theo sensor_id

```python
df_sr_info = df_doc_cam_bien.merge(df_thong_tin,on='sensor_id',how='left')
df_sr_info.head()
```

4.2 - Merge thêm maintenance_log theo machine_id

```python
df_day_du = df_sr_info.merge(df_bao_tri,on='machine_id',how='left')
df_day_du.head()
```
